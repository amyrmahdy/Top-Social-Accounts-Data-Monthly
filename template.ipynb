```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime
import os
import csv



# Set headers for HTTP requests:
headers = {
    'Accept-Encoding': 'gzip, deflate, sdch',
    'Accept-Language': 'en-US,en;q=0.8',
    'Upgrade-Insecure-Requests': '1',
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Cache-Control': 'max-age=0',
    'Connection': 'keep-alive',
}

# List of most-followed TikTok accounts
tiktok_url = 'https://en.wikipedia.org/wiki/List_of_most-followed_TikTok_accounts'



tiktok_result = requests.get(tiktok_url, headers = headers)

tiktok_soup = BeautifulSoup(tiktok_result.content, 'html5lib')
```

# Table

```python
tiktok_table = tiktok_soup.find("table",{"class":"wikitable sortable"})
```

# Columns

```python
tiktok_trs = tiktok_table.find_all("tr")
tiktok_zero_row = tiktok_trs[0].find_all("th")
tiktok_columns = []
for i in range(len(tiktok_zero_row)):
    tiktok_columns.append(tiktok_zero_row[i].text.strip())
print(tiktok_columns)
```

# Rows

```python
tiktok_rows = []
for i in range(1,51):
    tiktok_ith_row = tiktok_trs[i].find_all("td")
    tiktok_row = []
    tiktok_row.append(i)
    for i in range(len(tiktok_ith_row)):
        tiktok_row.append(tiktok_ith_row[i].text.strip())
    tiktok_rows.append(tiktok_row)
```

# DataFrame

```python
tiktok = pd.DataFrame(tiktok_rows, columns = tiktok_columns)
tiktok.drop(["Brandaccount"], axis = 1, inplace = True)
tiktok
```

# Save

```python
now = datetime.datetime.now()
filename = f"{now.strftime('%Y-%m')}/TikTok_top_50_{now.strftime('%Y-%m-%d_%H-%M-%S')}.csv"
os.makedirs(os.path.dirname(filename), exist_ok=True)
tiktok.to_csv(filename, index = False)
```
